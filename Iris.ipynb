## Import

```
import numpy as np
import tensorflow as tf
from keras.models import Sequential #for NN architecture
from keras.layers import Dense##for layer
from keras.utils import to_categorical
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

## Load DataSet

```
iris=load_iris()
X=preprocessing.scale(iris['data'])
Y=to_categorical(iris['target'])
```

## Splitting datset

```
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

## Build Model

```
model=Sequential()
model.add(Dense(20,input_dim=4,activation='relu'))
model.add(Dense(3,activation='softmax'))
model.summary()
```

```
model.compile(optimizer='Adam',
              loss='categorical_crossentropy', #as we have more than 2 classes or else binary
              metrics=['accuracy']
              )
```

## Train

```
model.fit(
    x_train,y_train,
    validation_data=(x_test,y_test),
    epochs=50,
    batch_size=32 # no. of samples per gradient update
)
```

```
model.fit(
    x_train,y_train,
    validation_data=(x_test,y_test),
    epochs=200,
    batch_size=32 # no. of samples per gradient update
)
```

```
model.fit(
    x_train,y_train,
    validation_data=(x_test,y_test),
    epochs=400,
    batch_size=32 # no. of samples per gradient update
)
```
